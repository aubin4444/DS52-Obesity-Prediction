# 4) Evaluating

```python
from sklearn.metrics import accuracy_score, make_scorer 
from sklearn.model_selection import cross_val_score

def cross_validation(model, X_train, y_train, n_folds=12):
    scoring = make_scorer(accuracy_score)
    scores = cross_val_score(model, X_train, y_train, cv=n_folds, scoring=scoring)
    mean_accuracy = np.mean(scores)
    print("Mean CV accuracy: ", mean_accuracy)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import StratifiedKFold

def print_confusion_matrix(confusion_matrix, class_labels):
    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_rows', None)
    pd.set_option('display.width', None)
    pd.set_option('display.expand_frame_repr', False)
    
    cm_df = pd.DataFrame(confusion_matrix, index=class_labels, columns=class_labels)
    print("Mean CV confusion matrix:")
    print(cm_df)

def cross_validation_with_confusion_matrix(model, X_train, y_train, n_folds=12):
    confusion_matrices = []
    scores = []

    class_labels = y_train.unique()  # Extract unique class labels

    for train_idx, val_idx in StratifiedKFold(n_splits=n_folds).split(X_train, y_train):
        X_fold_train, X_fold_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
        y_fold_train, y_fold_val = y_train.iloc[train_idx], y_train.iloc[val_idx]
        
        model.fit(X_fold_train, y_fold_train)
        y_pred = model.predict(X_fold_val)
        
        confusion_matrices.append(confusion_matrix(y_fold_val, y_pred , labels=class_labels))
        scores.append(accuracy_score(y_fold_val, y_pred))
        
    mean_accuracy = np.mean(scores)
    mean_confusion_matrix = np.mean(confusion_matrices, axis=0)
    
    print("Mean CV accuracy: ", mean_accuracy)
    #print("Mean CV confusion matrix: \n", mean_confusion_matrix)
    print_confusion_matrix(mean_confusion_matrix, class_labels)

# Usage example
cross_validation_with_confusion_matrix(classifier, X_train, y_train)
```
